```python
X = df.drop('Name', axis=1)
y = df['Name']

print(X)
print(y)

random =  42

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random)
```

---
***모델 기본 작동 순서***
- ***fit(*** X 학습 데이터, Y 학습 데이터 ***)***
- ***predict*** => return X_test에 대한 Y 예측
- ***accuracy_score(lib)*** => Y 테스트 데이터와 Y 예측 데이터를 비교하여 성능 측정

```python
def learn_something(model, xTrain, yTrain, xTest, yTest) :
    model.fit(xTrain, yTrain)
    y_pred = model.predict(xTest)
    accuracy = accuracy_score(yTest, y_pred)
    print(f"{model.__class__.__name__} : {accuracy}")
    return y_pred
```

---
***StandardScaler*** => 특성( Feature )의 값을 변환하는 라이브러리

스케일링 방법 
- 평균을 0, 표준편차를 1로 만듦 => 분포의 중심을 (0, 0)으로 설정
- 값의 범위를 정함
- ...

StandardScaler는 첫 번째 방법을 사용
- ***fit(X)*** => return StandardScaler()    학습된 스케일러 반환
- ***transform(X)*** => return scaled_X      스케일링 된 데이터
- ***fit_transform(X)*** => return scaled_X  fit(X)와 transform(X)를 같이 수행

학습 데이터로 ***fit_transform()*** 을 하고 test 데이터는 ***transsform()*** 만

```python
# 스케일링 적용
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

---
***KNN***

데이터에 인접한 k 개의 데이터의 결과를 따름

동점일 겨우
- 더 가까운 데이터의 결과 선택
- 임의( random ) 로 선택
- 동점이 생기지 않게 k 값 조정
- ...

```python
# 스케일링 적용 후 모델 학습 및 평가
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
y_pred_after = knn.predict(X_test_scaled)

print(f"{'predict':<17} {'test':<17} {'compare result'}")
for i in range(len(y_test)) :
    print(f"{y_pred_after[i] :<17} {list(y_test)[i] :<17} => {y_pred_after[i] == list(y_test)[i]}")

accuracy_after = accuracy_score(y_test, y_pred_after)
print("Accuracy after scaling:", accuracy_after)
```

***KNN에서 최적의 k값( n_neighbor )을 구하는 법 중 ... for 사용***

k에 따라서 결과가 달라지기 때문에 k의 값을 적절히 정하는 것이 중요

```python
# 변수 초기화
best_n_neighbors = 1
best_accuracy = 0

# n_neighbors 값을 1부터 20까지 반복하며 모델 성능 평가
for n in range(1, 21):
    # KNN 모델 생성
    knn = KNeighborsClassifier(n_neighbors=n)
    
    # 모델 훈련
    knn.fit(X_train_scaled, y_train)
    
    # 테스트 데이터에 대한 예측
    y_pred = knn.predict(X_test_scaled)
    
    # 정확도 계산
    accuracy = accuracy_score(y_test, y_pred)
    
    # 정확도가 가장 높은 n_neighbors 값 저장
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_n_neighbors = n
        
    if accuracy == 1 :
        break

# 최적의 n_neighbors 값 출력
print(f"최적의 n_neighbors 값: {best_n_neighbors}")
print(f"최고 정확도: {best_accuracy:.4f}")
```

3. 결측치 제거 && 불필요한 컬럼 제거 (Age인 경우 평균나이로)

    제거 방법

    - 유지 : 결측치를 그대로 유지하고 분석 방법을 조정

```python
X = df.drop("Survived", axis=1)
y = df["Survived"]

random = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random)

# 스케일링 적용
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# DT(Decision Tree)
DT_pred_y = learn_something(DecisionTreeClassifier(random_state=random), X_train, y_train, X_test, y_test)

# RF(Random Forest)
RF_pred_y = learn_something(RandomForestClassifier(random_state=random), X_train, y_train, X_test, y_test)

# SVC(Support Vector Classifier)
SVC_pred_y = learn_something(SVC(random_state=random), X_train, y_train, X_test, y_test)

# LR(LogisticRegression)
LR_pred_y = learn_something(LogisticRegression(random_state=random, max_iter=300), X_train, y_train, X_test, y_test)

# KNN(K-Nearest Neighbor)
KNN_pred_y = learn_something(KNeighborsClassifier(n_neighbors=3), X_train_scaled, y_train, X_test_scaled, y_test)

# def learn_something(model, xTrain, yTrain, xTest, yTest) :
#     model.fit(xTrain, yTrain)
#     y_pred = model.predict(xTest)
#     accuracy = accuracy_score(yTest, y_pred)
#     print(f"{model.__class__.__name__} : {accuracy}")
```
