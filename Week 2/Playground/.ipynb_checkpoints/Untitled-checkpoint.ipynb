```python
import pandas as pd


df = pd.read_csv("https://github.com/MyungKyuYi/AI-class/blob/main/titanic.csv?raw=true")
df.head()
```

```python
df["Survived"].value_counts()
```

```python
df.isnull().sum()
```

```python
# Label을 숫자로 변환
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for col in df.columns :
    df[col] = label_encoder.fit_transform(df[col])

# 특정 column 삭제
# df.drop(columns=[])

# Name & Ticket & Cabin & Embarked 삭제
df = df.drop(columns=["Name", "Ticket", "Cabin", "Embarked"])

# 평균으로 대체
# mid = df['Column Name'].mean()
# df['Column Name'] = df['Column Name'].fillna(mid)

# Age의 빈 값을 평균으로 대체
mid = df['Age'].mean()
df['Age'] = df['Age'].fillna(mid)

df.head()
```

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score

X = df.drop("Survived", axis=1)
y = df["Survived"]

random = 42

# KNN을 위한 데이터 스케일링
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random)

# DT(Decision Tree)
model = DecisionTreeClassifier(random_state=random)

# RF(Random Forest)
# model = RandomForestClassifier(random_state=random)

# SVC(Support Vector Classifier)
# model = SVC(random_state=random)

# LR(LogisticRegression)
# model = LogisticRegression(random_state=random, max_iter=200)

# KNN(K-Nearest Neighbor)
# model = KNeighborsClassifier(n_neighbors=3)

model.fit(X_train, y_train)
y_predict = model.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)

print(f"result: {accuracy:.4f}")
```
