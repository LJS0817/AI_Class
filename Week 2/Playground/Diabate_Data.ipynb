```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score

# df = pd.read_csv("https://github.com/MyungKyuYi/AI-class/blob/main/diabetes.csv?raw=true")
df = pd.read_csv("C:/Users/admin/Desktop/Homework/AI/AI_Class/Week 2/Data/diabetes.csv")
df.head()
```

```python
random = 42

X = df.drop("Outcome", axis=1)
y = df["Outcome"]

# KNN을 위한 데이터 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random)

print(X_train)
print(X_test)

X_train, X_test, y_train, y_test = train_test_split(df.drop("Outcome", axis=1), df["Outcome"], test_size=0.2, random_state=random)

print(scaler.fit_transform(X_train))
print(scaler.fit_transform(X_test))
```
